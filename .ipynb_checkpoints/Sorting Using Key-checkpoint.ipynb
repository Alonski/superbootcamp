```python
import csv
with open("countries_of_the_world/cow.csv") as f:
    countries = list(csv.DictReader(f))
len(countries)
```

```python
def get_population(d):
    return d['name'][0], -int(d['population'])

countries.sort(key=get_population)

[(c['name'], c['population']) for c in countries]
```
